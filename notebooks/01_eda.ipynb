```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/deliveries_raw.csv")

print("SHAPE:", df.shape)
print("\nCOLUMNS:", df.columns.tolist())
print("\nDTYPES:\n", df.dtypes)
print("\nHEAD:\n", df.head(3))
print("\nNULLS:\n", df.isnull().sum())
```

```python
object_cols = df.select_dtypes(include='object').columns
for col in object_cols:
    nan_count = (df[col].astype(str).str.strip() == "NaN").sum()
    if nan_count > 0:
        print(f"{col}: {nan_count} hidden NaN strings")
```

```python
cat_cols = ['Weatherconditions', 'Road_traffic_density', 'Type_of_order', 
            'Type_of_vehicle', 'Festival', 'City']
for col in cat_cols:
    print(f"\n{col} — unique values:")
    print(df[col].value_counts(dropna=False))
```

```python
print("Delivery_person_Age sample:", df['Delivery_person_Age'].head(10).tolist())
print("\nDelivery_person_Ratings sample:", df['Delivery_person_Ratings'].head(10).tolist())
print("\nmultiple_deliveries sample:", df['multiple_deliveries'].head(10).tolist())
print("\nTime_taken(min) sample:", df['Time_taken(min)'].head(10).tolist())
```

```python
print("Zero coords — restaurant:", (df['Restaurant_latitude'] == 0).sum())
print("Zero coords — delivery:", (df['Delivery_location_latitude'] == 0).sum())
print("Negative lat:", (df['Restaurant_latitude'] < 0).sum())
print("Lat range:", df['Restaurant_latitude'].min(), "to", df['Restaurant_latitude'].max())
print("Lng range:", df['Restaurant_longitude'].min(), "to", df['Restaurant_longitude'].max())
```
